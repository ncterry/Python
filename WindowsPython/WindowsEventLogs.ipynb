```python
'''
Windows event logs: 
Python script that reads and displays the latest # of events from Windows Event Logs

Accessing the "Security" log might require administrator privileges. 

EXAMPLE
server = None  # None for the local machine
logtype = ["System", "Application", "Security"]
event_count = 5

windowsEventLogs(server, logtype[0], event_count)
'''
import win32evtlog  # Import the required module
import win32evtlogutil  # Import the win32evtlogutil module

def windowsEventLogs(server, logtype, event_count):

    # Open the event log
    handle = win32evtlog.OpenEventLog(server, logtype)

    # Read the latest 10 events (change the number to retrieve more events)
    flags = win32evtlog.EVENTLOG_BACKWARDS_READ | win32evtlog.EVENTLOG_SEQUENTIAL_READ
    events = win32evtlog.ReadEventLog(handle, flags, 0)

    count = 0
    for event in events:
        # How many events do we want printed
        if count >= event_count:
            break

        # Event information
        print(f"Event ID: {event.EventID & 0x1FFFFFFF}")  # Mask out the higher bits
        print(f"Event Type: {event.EventType}")
        print(f"Time Generated: {event.TimeGenerated}")
        print(f"Source Name: {event.SourceName}")
        print(f"Category: {event.EventCategory}")

        # Event message
        msg = win32evtlogutil.SafeFormatMessage(event, logtype)
        print(f"Message: {msg}")

        print("-" * 50)
        count += 1

    # Close the event log
    win32evtlog.CloseEventLog(handle)

```

```python
server = None  # None for the local machine
# The log you want to read; common types are "System", "Application", "Security"
#logtype = "System"  
#logtype = "Application"
#logtype = "Security"
logtype = ["System", "Application", "Security"]
# How many events do you want to be printed
event_count = 5

windowsEventLogs(server, logtype[2], event_count)
```
