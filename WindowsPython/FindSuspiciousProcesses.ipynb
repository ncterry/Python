```python
'''
Analyzing running processes for suspicious activity:

Using the psutil library, you can check for suspicious processes by matching process names, 
executable paths, or command line arguments against a list of known suspicious items.
'''
import psutil
def findSuspiciousNames(suspicious_process_names):

    for process in psutil.process_iter(['name', 'exe', 'cmdline']):
        process_info = process.info
        print(process)
        #Ex:
        #psutil.Process(pid=4, name='System', status='running', started='2023-04-12 01:01:13')
        #psutil.Process(pid=8, name='svchost.exe', status='running', started='2023-04-12 01:01:20')
        
        if process_info['name'] in suspicious_process_names or process_info['exe'] in suspicious_process_names:
            print(f"Suspicious process found: {process_info['name']} (PID: {process.pid})")
```
