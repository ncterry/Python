```python
#pip install pywin32
import win32evtlog
import win32evtlogutil
import win32con
import re
'''
This function searches the specified Windows Event Log (log_type) for 
occurrences of the given port number. By default, the function looks 
at the 'System' log. You can modify the log_type parameter to search 
in other logs, such as 'Application' or 'Security'.

Please note that this script might not cover all cases where a port 
is used, as not every usage of a port generates an event log entry. 
Additionally, you may need to run the script with administrative privileges 
to access certain event logs.
'''
def search_event_logs_for_port(port, log_type='System'):
    server = None  # None means the local computer
    log = log_type
    flags = win32evtlog.EVENTLOG_BACKWARDS_READ | win32evtlog.EVENTLOG_SEQUENTIAL_READ

    # Compile regex pattern to search for port
    port_pattern = re.compile(f'\\b{port}\\b')

    # Open the event log
    log_handle = win32evtlog.OpenEventLog(server, log)

    # Read events
    while True:
        events = win32evtlog.ReadEventLog(log_handle, flags, 0)
        if not events:
            break
        for event in events:
            # EventMessageFile provides access to the message text associated with an event
            event_message = win32evtlogutil.SafeFormatMessage(event, log_type)
            if port_pattern.search(event_message):
                print(f"Event ID: {event.EventID}")
                print(f"Source: {event.SourceName}")
                print(f"Time: {event.TimeGenerated}")
                print(f"Event Message: {event_message}\n")

    win32evtlog.CloseEventLog(log_handle)

```
